```python
from mxlmodels import Simulator, get_yokota1985, plot

res = Simulator(get_yokota1985()).simulate(10).get_result().unwrap_or_err()

fig, (ax1, ax2) = plot.two_axes(figsize=(12, 6))
_ = plot.lines(res.variables.iloc[:, :10], ax=ax1)
_ = plot.lines(res.fluxes.iloc[:, :10], ax=ax2)
```

```python
res = (
    Simulator(get_yokota1985())
    .update_variables(
        {
            "glycolate": 0,
            "glyoxylate": 0,
            "glycine": 0,
            "serine": 0,
            "hydroxypyruvate": 0,
            "H2O2": 0,
        }
    )
    .simulate(1e-1)
    .get_result()
    .unwrap_or_err()
)

fig, (ax1) = plot.one_axes(figsize=(4, 4))
_ = plot.lines(res.variables.iloc[:, :10], ax=ax1)
ax1.set(xlabel="Time / h", ylabel=r"Concentration / $\mu\ mol\ mg^{-1}\ Chl\ h^{-1}$")
fig.savefig("yokota1980-fig6a.svg")
```

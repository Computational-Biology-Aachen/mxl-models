```python
from mxlmodels import Simulator, get_dynamic_enterobactin, plot

res = Simulator(get_dynamic_enterobactin()).simulate(0.2).get_result().unwrap_or_err()

fig, (ax1, ax2, ax3) = plot.grid_layout(3, n_cols=3)
_ = plot.lines(res.get_variables(include_derived_variables=False).iloc[:, :10], ax=ax1)
_ = plot.lines(
    res.get_args(
        include_variables=False,
        include_derived_variables=True,
        include_reactions=False,
    ).iloc[:, :10],
    ax=ax2,
)
_ = plot.lines(res.fluxes.iloc[:, :10], ax=ax3)
```

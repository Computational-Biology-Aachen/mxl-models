```python
import numpy as np
from mxlpy import Model
from mxlpy import Simulator
import matplotlib.pyplot as plt
from mxlmodels import get_salvatori2022
```

```python
concs=[]
fluxes=[]
for x in ["Eiko", "Minn"]:
    m = get_salvatori2022(load=x)
    s = Simulator(m)
    s.simulate(1)
    c,v = s.get_result().unwrap_or_err()
    y0 = dict(c.iloc[-1])
    m = get_salvatori2022(load=x)
    s = Simulator(m)
    s.update_variables(y0)
    for i in range(1, 60):
        if i == 1:
            s.update_parameter('PAR', 0)
        elif i % 2 == 0:
            s.update_parameter('PAR', 520)
        else:
            s.update_parameter('PAR', 780)
        s.simulate(i, steps=1000)

    c, v= s.get_result().unwrap_or_err()
    concs.append(c)
    fluxes.append(v)
```

```python
font = {'size'   : 11}
plt.rc('font', **font)

fig, axs = plt.subplots(nrows = 3, sharex = True, figsize = (7.99, 6.18), layout = 'constrained')

#Format for all axs
for i in range(len(axs)):
    axs[i].tick_params(bottom = True, top = True, right = True, left = True, direction = 'in')

axs[0].set_ylabel('ETR [µmol m⁻² s⁻¹]')
axs[0].set_ylim(0, 110)
axs[0].set_yticks([0, 50, 100])
axs[0].plot(concs[0].index, fluxes[0]['ETR_out'], color="forestgreen")
axs[0].plot(concs[1].index, fluxes[1]['ETR_out'], color="lightgreen")

#Color bar for PAR
flag_Dark = 0
flag_520 = 0
flag_780 = 0
for y_pos in [1.01, 1.02, 1.012]:
    
    for i in range(len(s.simulation_parameters)):
        
        if s.simulation_parameters[i]['PAR'] == 0:
            if flag_Dark == 0:
                axs[0].plot([i, i+1], [y_pos, y_pos], clip_on = False, transform=axs[0].get_xaxis_transform(), color = "gray", linewidth = 1.5, label = 'Dark')
                flag_Dark = 1
            else:
                axs[0].plot([i, i+1], [y_pos, y_pos], clip_on = False, transform=axs[0].get_xaxis_transform(), color = "gray", linewidth = 1.5)
                
        elif s.simulation_parameters[i]['PAR'] == 520:
            if flag_520 == 0:
                axs[0].plot([i, i+1], [y_pos, y_pos], clip_on = False, transform=axs[0].get_xaxis_transform(), color = "yellow", linewidth = 1.5, label = '520 PPFD')
                flag_520 = 1
            else:
                axs[0].plot([i, i+1], [y_pos, y_pos], clip_on = False, transform=axs[0].get_xaxis_transform(), color = "yellow", linewidth = 1.5)
                
        elif s.simulation_parameters[i]['PAR'] == 780:
            if flag_780 == 0:
                axs[0].plot([i, i+1], [y_pos, y_pos], clip_on = False, transform=axs[0].get_xaxis_transform(), color = "orange", linewidth = 1.5, label = '780 PPFD')
                flag_780 = 1
            else:
                axs[0].plot([i, i+1], [y_pos, y_pos], clip_on = False, transform=axs[0].get_xaxis_transform(), color = "orange", linewidth = 1.5)

axs[0].legend(loc = (0.46, 1.06), ncol = 3)
            
axs[1].set_ylabel('A [µmol m⁻² s⁻¹]')
axs[1].set_ylim(0, 20)
axs[1].set_yticks([i * 5 for i in range(5)])
axs[1].plot(concs[0].index, concs[0]['A'], color="forestgreen")
axs[1].plot(c.index, concs[1]['A'], color="lightgreen")

axs[2].set_ylabel('NPQ [µmol m⁻² s⁻¹]')
axs[2].set_ylim(0, 2.5)
axs[2].set_yticks([i * 0.5 for i in range(6)])
axs[2].plot(fluxes[0].index, fluxes[0]['NPQ'], color="forestgreen", label="Eiko")
axs[2].plot(c.index, fluxes[1]['NPQ'], color="lightgreen", label="MinnGold")
axs[2].set_xlim(0, 60)
axs[2].set_xlabel('time [min]')
axs[2].legend()
plt.savefig(fname = 'Eiko_own.png', dpi = 100.0, transparent = False, facecolor = 'white')
plt.savefig("salvatori2022_fig3.svg")
```

```python
concs = []
fluxes = []

periods = [0.5, 1.0, 4.0]   # fluctuation periods in seconds
T_end = 60                  # total simulation time (seconds)
dt = 0.1                    # integration step (seconds)

# ----------------------------------------------------
# 1. CONSTANT LIGHT CONTROL (650 µmol m⁻² s⁻¹ for 60 s)
# ----------------------------------------------------
m = get_salvatori2022()
s = Simulator(m)

s.update_variables(y0)

t = 0.0
while t < T_end:
    s.update_parameter("PAR", 650)   # constant light
    s.simulate(t + dt, steps=100)
    t += dt

c, v = s.get_result().unwrap_or_err()
concs.append(c)
fluxes.append(v)

# ----------------------------------------------------
# 2. FLUCTUATING LIGHT CASES
# ----------------------------------------------------
for period in periods:

    m = Salvatori2022()
    s = Simulator(m)

    s.update_variables(y0)

    t = 0.0
    while t < T_end:

        # Square-wave switching every "period" seconds
        if int(t // period) % 2 == 0:
            s.update_parameter("PAR", 520)
        else:
            s.update_parameter("PAR", 780)

        s.simulate(t + dt, steps=100)
        t += dt

    c, v = s.get_result().unwrap_or_err()
    concs.append(c)
    fluxes.append(v)
```

# Hahn1987

[https://doi.org/10.1093/oxfordjournals.aob.a087432](https://doi.org/10.1093/oxfordjournals.aob.a087432)

```python
import string

import matplotlib.pyplot as plt
import numpy as np

from mxlmodels import Simulator, get_hahn1987
```

## Figure 2

```python
res_fig2 = {}

for time in [4 * 60, 80 * 60]:
    res_fig2[str(time)] = (
        Simulator(
            get_hahn1987().update_variables(
                {
                    "RuBP": 0,
                    "TP": 0,
                    "HP": 0,
                    "TPGA": 0,
                    "E4P": 0,
                    "S7P": 0,
                    "Ru5P": 0,
                    "GA": 0,
                    "PGl": 0,
                    "Gl": 0,
                    "Gn": 0,
                    "Gx": 0,
                }
            )
        )
        .simulate(time, steps=int(1e4))
        .get_result()
        .unwrap_or_err()
        .get_combined()
    )

fig, axs = plt.subplots(nrows=2, figsize=(5, 10))

to_plot = {"TP": 1, "Gl": 1, "RuBP": 1, "HP": 0.67, "PGA": 0.5, "PGl": 0.2, "NH3": 0.4}

for i, (time, result) in enumerate(res_fig2.items()):
    add_label = 0.1 if time == str(4 * 60) else 1.0
    for var, factor in to_plot.items():
        var_res = result[var] * factor
        axs[i].plot(var_res.index / 60, var_res, label=var, color="black")
        axs[i].text(
            var_res.index[-1] / 60 + add_label,
            var_res.iloc[-1],
            var,
            va="center",
            ha="left",
        )

for i, ax in enumerate(axs):
    ax.set_ylim(0, 1)
    ax.set_xlabel("Time (min)")
    ax.set_yticks(np.linspace(0, 1.0, 11))
    ax.text(
        0.02,
        0.98,
        string.ascii_uppercase[i],
        transform=ax.transAxes,
        fontsize=10,
        va="top",
    )

axs[0].set_ylabel("Concentration (mM)")
axs[0].set_xticks(np.linspace(0, 4.0, 9))
axs[0].set_xlim(0, 4.6)

axs[1].set_xticks(np.linspace(0, 80, 9))
axs[1].set_xlim(0, 90)

plt.tight_layout()
plt.show()
```

## Figure 3

```python
res_fig3 = {}

for ca_val in [0.05, 4.5]:
    res_fig3[str(ca_val)] = (
        Simulator(get_hahn1987().update_parameter("Ca", ca_val))
        .simulate(2 * 60, steps=int(1e4))
        .get_result()
        .unwrap_or_err()
        .get_combined()
    )

fig, axs = plt.subplots(nrows=2, figsize=(5, 10))

to_plot = {
    "RuBP": {"0.05": 0.25, "4.5": 1},
    "PGA": {"0.05": 0.5, "4.5": 0.25},
    "TP": {"0.05": 0.5, "4.5": 0.33},
    "PGl": {"0.05": 0.5, "4.5": 1},
    "Gl": {"0.05": 0.57, "4.5": 1},
    "Gn": {"0.05": 0.5, "4.5": 0.8},
    "NH3": {"0.05": 0.67, "4.5": 0.8},
    "PAn": {"0.05": 0.02, "4.5": 0.005},
}

for i, (ca_val, result) in enumerate(res_fig3.items()):
    for var, factor_dict in to_plot.items():
        factor = factor_dict[ca_val]
        var_res = result[var] * factor
        axs[i].plot(var_res.index / 60, var_res, label=var, color="black")
        axs[i].text(
            var_res.index[-1] / 60 + 0.05, var_res.iloc[-1], var, va="center", ha="left"
        )

for i, ax in enumerate(axs):
    ax.set_xlabel("Time (min)")
    ax.set_ylabel("Concentration (mM)")
    ax.set_xlim(0, 2.35)
    ax.set_xticks(np.linspace(0, 2, 9))
    ax.set_ylim(0, 1.05)
    ax.set_yticks(np.linspace(0, 1, 11))
    ax.text(
        0.98,
        0.98,
        string.ascii_uppercase[i],
        transform=ax.transAxes,
        fontsize=10,
        va="top",
        ha="right",
    )

plt.tight_layout()
plt.show()
```

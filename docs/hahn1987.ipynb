```python
from mxlmodels import Simulator, get_hahn1987, plot
import matplotlib.pyplot as plt
```

```python
res = (
    Simulator(get_hahn1987())
    .simulate(1)
    .get_result()
    .value
)

variables, fluxes = res
fig, ax = plt.subplots()
plot.lines(variables, ax=ax)

```

```python
from mxlpy import Simulator, plot
import matplotlib.pyplot as plt

res = (
    Simulator(get_hahn1987().update_variables(
        {
            "RuBP":0,
            "TP": 0,
            "HP": 0,
            "TPGA": 0,
            "E4P": 0,
            "S7P": 0,
            "Ru5P": 0,
            "GA": 0,
            "PGl": 0,
            "Gl": 0,
            "Gn": 0,
            "Gx": 0
        }
    ))
    .simulate(4*60)
    .get_result()
    .value
)

variables, fluxes = res
fig, ax = plt.subplots()
plot.lines(variables.TP, ax=ax)
plot.lines(variables.Gl, ax=ax)
plot.lines(variables.RuBP, ax=ax)
plot.lines(variables.HP*0.67, ax=ax)
plot.lines(variables.PGA*0.5, ax=ax)
plot.lines(variables.PGl*0.2, ax=ax)
plot.lines(variables.NH3*0.4, ax=ax)
```
